```python
def find_root(f, a, b, e):
    if f(a) > 0 or f(b) < 0:
        return None
    middle = (a + b) / 2
    if abs(f(middle)) < e:
        return middle
    else:
        if f(middle) > 0:
            return find_root(f, a, middle, e)
        else:
            return find_root(f, middle, b, e)
        
def f(x):
    return x ** 2 - 2

def g(x):
    return x ** x + 10

import matplotlib.pyplot as plt
import numpy as np
def plot(function, a, b, e):
    x = np.arange(a, b, e)
    y = function(x)
    plt.axhline(y=0, color='k')
    plt.plot(x, y)
    plt.grid(True)
    plt.show()

print(find_root(g, 0, 5, 0.0000000000001))
plot(g, 0, 5, 0.0001)
```
